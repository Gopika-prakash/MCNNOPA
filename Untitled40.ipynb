```python
import torch
import numpy as np
import cv2
from torchvision import datasets, transforms
import random



# Fourier data augmentation function
def fourier_augment(image, alpha=0.1):
    """
    Apply Fourier Data Augmentation by perturbing the image in the frequency domain.
    alpha: Determines the intensity of perturbation.
    """
    # Convert image to numpy array and switch to the frequency domain
    image_np = image.numpy().transpose(1, 2, 0)  # Convert to HWC format for processing
    image_np = cv2.cvtColor(image_np, cv2.COLOR_RGB2GRAY)  # Convert to grayscale (if needed)

    # Perform Fourier transform
    f_image = np.fft.fft2(image_np)
    f_image_shifted = np.fft.fftshift(f_image)  # Shift zero frequency component to center

    # Apply perturbation (random noise in the frequency domain)
    rows, cols = image_np.shape
    crow, ccol = rows // 2 , cols // 2  # Center
    mask = np.ones((rows, cols), np.uint8)

    # Randomly mask a small area around the center frequencies (low frequencies)
    mask[crow-int(alpha*crow):crow+int(alpha*crow), ccol-int(alpha*ccol):ccol+int(alpha*ccol)] = 0

    f_image_shifted = f_image_shifted * mask
    f_image_unshifted = np.fft.ifftshift(f_image_shifted)

    # Perform Inverse Fourier transform to get back the spatial domain image
    img_back = np.fft.ifft2(f_image_unshifted)
    img_back = np.abs(img_back)  # Take the magnitude

    # Normalize back to image format (0-1 scale)
    img_back = (img_back - np.min(img_back)) / (np.max(img_back) - np.min(img_back))

    # Convert back to tensor
    img_back = torch.tensor(img_back, dtype=torch.float32).unsqueeze(0)  # Add channel dimension back (1 x H x W)

    # Convert grayscale back to 3-channel image
    img_back = img_back.expand(3, -1, -1)  # Duplicate across RGB channels

    return img_back

# Custom transform that includes Fourier Data Augmentation
class FourierTransform:
    def __call__(self, img):
        # Apply Fourier augmentation to a subset of images randomly
        if random.random() < 0.5:  # Apply to 50% of the images
            img = fourier_augment(img)
        return img

# Step 1: Load and Preprocess the Dataset with Data Augmentation
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    FourierTransform(),  # Add Fourier augmentation to the pipeline
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # Normalization for RGB images
])
transform1 = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    FourierTransform(),  # Add Fourier augmentation to the pipeline
      # Normalization for RGB images
])
odataset = datasets.ImageFolder(root=dataset_path,transform=transform1)
dataset = datasets.ImageFolder(root=dataset_path, transform=transform)

# Calculate and display dataset size
train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size

print(f'Total Dataset Size: {len(dataset)} images')
print(f'Training Dataset Size: {train_size} images')
print(f'Validation Dataset Size: {val_size} images')
print(f'Testing Dataset Size: {test_size} images')
```

```python
# Step 3: Enhanced ABiLSTM with Attention Mechanism
class Attention(nn.Module):
    def __init__(self, hidden_size):
        super(Attention, self).__init__()
        self.attention_weights = nn.Parameter(torch.randn(hidden_size * 2, 1))

    def forward(self, lstm_output):

        attention_scores = torch.tanh(torch.matmul(lstm_output, self.attention_weights)).squeeze(-1)
        attention_weights = torch.softmax(attention_scores, dim=-1)  # softmax over the sequence_length (last dim)
        context_vector = torch.bmm(attention_weights.unsqueeze(1), lstm_output).squeeze(1)
        return context_vector
```

```python
class ABiLSTMWithAttention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes, dropout_prob=0.5):
        super(ABiLSTMWithAttention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, bidirectional=True)
        self.attention = Attention(hidden_size)
        self.fc = nn.Linear(hidden_size * 2, num_classes)
        self.dropout = nn.Dropout(dropout_prob)
        self.batch_norm = nn.BatchNorm1d(hidden_size * 2)  # BatchNorm layer

    def forward(self, x):
        h0 = torch.zeros(2 * self.lstm.num_layers, x.size(0), self.lstm.hidden_size).to(x.device)
        c0 = torch.zeros(2 * self.lstm.num_layers, x.size(0), self.lstm.hidden_size).to(x.device)

        x = x.unsqueeze(1)
        out, _ = self.lstm(x, (h0, c0))
        out = self.dropout(out)
        out = self.attention(out)
        out = self.batch_norm(out)  # Apply batch normalization
        out = self.fc(out)
        return out
```

```python
# Initialize the ABiLSTM model
input_size = 50176
hidden_size = 256
num_layers = 2
num_classes = 25

# Call the correct class name ABiLSTMWithAttention
abilstm = ABiLSTMWithAttention(input_size=input_size, hidden_size=hidden_size, num_layers=num_layers, num_classes=num_classes)

# Display the summary of ABiLSTM
print("ABiLSTM Summary:")
print(abilstm)

# Step 5: Training and Evaluation
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
feature_extractor = feature_extractor.to(device)
abilstm = abilstm.to(device)
# Assuming manual_weights should be a tensor of weights for each class
# Replace this with your actual weight values
manual_weights = torch.tensor([1.0, 0.5, 0.7, 0.3, 0.9,
                               0.6, 0.8, 0.4, 0.2, 1.1,
                               0.5, 0.7, 0.7, 0.9, 0.6,
                               0.8, 0.4, 0.2, 1.1, 1.0,
                               0.5, 0.7, 0.3, 0.9, 0.6])

# Move manual_weights to the device
manual_weights = manual_weights.to(device)

class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = nn.CrossEntropyLoss(weight=self.alpha, reduction='none')(inputs, targets)
        pt = torch.exp(-ce_loss)  # Get probabilities
        focal_loss = ((1 - pt) ** self.gamma) * ce_loss

        if self.reduction == 'mean':
            return torch.mean(focal_loss)
        elif self.reduction == 'sum':
            return torch.sum(focal_loss)
        else:
            return focal_loss

# Use Focal Loss instead of CrossEntropyLoss
criterion = FocalLoss(alpha=manual_weights).to(device)
```

```python

#Implement in sOA as additional import numpy as np
import pickle
import os
import torch
import torch.optim as optim
import random

# Assumed constants and variables
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
population_size = 20
max_iters = 15  # Can be increased
S_min, S_max = 1e-6, 1e-1  # Learning rate range
K1, K2, K3 = 0.5, 0.05, 2.0
learning_rate_path = "/content/drive/MyDrive/best_learning_rate.txt"  # Path to save the best learning rate
save_path = "/content/drive/MyDrive/optimization_progress.pkl"  # Path to save the progress
# Temporary learning rate for the optimizer
initial_learning_rate = 1e-5

# Define optimizer with the model and initial learning rate
optimizer = optim.Adam(abilstm.parameters(), lr=initial_learning_rate)
# Random seed for reproducibility
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)
import numpy as np
import torch



# Function to initialize population
def initialize_population(pop_size, S_min, S_max):
    return S_min + np.random.rand(pop_size) * (S_max - S_min)

# Fitness function based on validation loss
def fitness_function(learning_rate):
    # Set the learning rate for the optimizer
    optimizer = optim.Adam(abilstm.parameters(), lr=learning_rate)

    # Train the model for a few epochs to evaluate fitness
    num_epochs = 5  # Short training for fitness evaluation
    running_loss = 0.0
    for epoch in range(num_epochs):
        feature_extractor.train()
        abilstm.train()
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            features = feature_extractor(images)
            outputs = abilstm(features)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

    avg_loss = running_loss / (num_epochs * len(train_loader))
    return avg_loss  # Minimize this value

# Function to update position (local and global search)
def update_position(S, best_position, food_position, Q, R, temperature, mode='exploration'):
    if mode == 'exploration':
        new_position = S + K2 * (np.exp(-fitness_function(food_position)) / np.exp(-fitness_function(best_position))) * (S_max - S_min) * R + S_min
    else:  # Exploitation
        if temperature > 0.6:  # Hot (move towards food)
            new_position = food_position + K3 * temperature * R * (food_position - S)
        else:  # Cold (mating or fighting mode)
            if Q > 0.25:  # Mating
                new_position = S + K3 * np.exp(-fitness_function(food_position) / fitness_function(S)) * R * (Q * best_position - S)
            else:  # Fighting
                new_position = S + K3 * np.exp(-fitness_function(best_position) / fitness_function(S)) * R * (Q * best_position - S)
    return np.clip(new_position, S_min, S_max)  # Keep within range

# Function to save model, optimizer, and progress state
def save_progress(iteration, population, fitness_values, model, optimizer):
    state = {
        'iteration': iteration,
        'population': population,
        'fitness_values': fitness_values,
        'model_state_dict': model.state_dict(),  # Save model state
        'optimizer_state_dict': optimizer.state_dict()  # Save optimizer state
    }
    with open(save_path, 'wb') as f:
        pickle.dump(state, f)
    print(f"Progress saved at iteration {iteration + 1}.")

# Function to load model, optimizer, and progress state
def load_progress(model, optimizer):
    if os.path.exists(save_path):
        with open(save_path, 'rb') as f:
            state = pickle.load(f)
        model.load_state_dict(state['model_state_dict'])  # Load model state
        optimizer.load_state_dict(state['optimizer_state_dict'])  # Load optimizer state
        print(f"Progress loaded from iteration {state['iteration'] + 1}.")
        return state['iteration'], state['population'], state['fitness_values']
    else:
        return 0, None, None  # No saved progress found

# Function to execute the Snake Optimization algorithm with save functionality after each iteration
def snake_optimization_with_save(population, fitness_values, max_iters, model, optimizer, start_iter=0):
    print("Initial Population and Fitness Values:")
    for i in range(population_size):
        print(f"Learning Rate: {population[i]:.6f}, Fitness: {fitness_values[i]:.6f}")

    for iteration in range(start_iter+1, max_iters):
        print(f"\nIteration {iteration + 1}/{max_iters}")

        # Calculate temperature and food quantity
        temperature = np.exp(-iteration / max_iters)
        Q = K1 * np.random.rand() * (1 - iteration / max_iters)

        # Separate population into males and females
        N_m = population_size // 2
        N_f = population_size - N_m
        male_population = population[:N_m]
        female_population = population[N_m:]

        # Calculate best individuals
        best_male = male_population[np.argmin(fitness_values[:N_m])]
        best_female = female_population[np.argmin(fitness_values[N_m:])]

        # Update positions for males and females
        R = np.random.rand()
        for i in range(N_m):
            population[i] = update_position(male_population[i], best_male, best_female, Q, R, temperature, mode='exploration')
        for i in range(N_m, population_size):
            population[i] = update_position(female_population[i - N_m], best_female, best_male, Q, R, temperature, mode='exploitation')

        # Recalculate fitness
        fitness_values = np.array([fitness_function(p) for p in population])

        # Print updated population and fitness values
        for i in range(population_size):
            print(f"Learning Rate: {population[i]:.6f}, Fitness: {fitness_values[i]:.6f}")

        # Save progress after each iteration
        save_progress(iteration, population, fitness_values, model, optimizer)

    return population, fitness_values

# Load saved progress or start from scratch
start_iter, saved_population, saved_fitness = load_progress(abilstm, optimizer)

if saved_population is not None:
    # Continue from saved state
    population = saved_population
    fitness_values = saved_fitness
else:
    # Initialize the population
    population = initialize_population(population_size, S_min, S_max)

    # Initial fitness calculation
    fitness_values = np.array([fitness_function(p) for p in population])

# Run Snake Optimization with saving after each iteration
optimized_population, optimized_fitness = snake_optimization_with_save(population, fitness_values, max_iters, abilstm, optimizer, start_iter) # Added abilstm and optimizer

# Once all iterations are complete, save the best learning rate
best_learning_rate = optimized_population[np.argmin(optimized_fitness)]
print(f"Optimized Learning Rate: {best_learning_rate:.6f}")

# Save the best learning rate to a file
with open(learning_rate_path, 'w') as f:
    f.write(str(best_learning_rate))
```
